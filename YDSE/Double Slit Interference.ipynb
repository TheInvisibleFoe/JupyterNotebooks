```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.animation as anim
# np.set_printoptions(threshold=sys.maxsize)
x_l = 20
y_l = 5
points = 1000
x = np.linspace(0,x_l,points)
y = np.linspace(-y_l,y_l,points)
gridx , gridy = np.meshgrid(x,y)
```

```python
d = 2
A = 50
omg = 5
k=5
dist_y_1 = gridy+d
dist_y_2 = gridy-d
```

```python
t=0
dist_1 = np.sqrt(gridx**2+dist_y_1**2)
dist_2 = np.sqrt(gridx**2+dist_y_2**2)
def wave1(t,dist):
    return A * np.sin(k*dist - omg* t)
def wave2(t,dist):
    return A * np.sin(k*dist - omg * t)
def wave(t,dist1,dist2):
    return wave1(t,dist1)+wave2(t,dist2)
```

```python
fig,axes = plt.subplots(figsize = (16,9))
axes.set_xlim(0,19)
axes.set_ylim(-4.9,4.9)
axes.scatter(gridx,gridy,c=wave(0,dist_1,dist_2),cmap = "Blues")
```
