# Finite Difference Method

We discretize Laplace's equations in 2 dimensions, using the three point formula in two dimensions

$$ \boxed{\dfrac{\partial^2 V}{\partial x^2} + \dfrac{\partial^2 V}{\partial y^2} = 0}{} $$

We discretize the equation in two dimension 

$$ \dfrac{V(x + \Delta x,y) + V(x - \Delta x,y) + V(x,y)}{\Delta x} +\dfrac{V(x, y + \Delta y) + V(x, y - \Delta y) + V(x,y)}{\Delta y}  =  0  $$

Setting $\Delta x = \Delta y$  we get the following equation,

$$ \boxed{V(x,y) = \dfrac{1}{4}\left[V(x + \Delta x,y) + V(x - \Delta x,y) + V(x, y + \Delta x) + V(x, y  - \Delta x)\right] }$$

````python
import numpy as np
import matplotlib.pyplot as plt
from numba import jit
````

````python
edge = np.linspace(-1,1,300)
xv, yv = np.meshgrid(edge, edge)
````

## Boundary Conditions

We need 4 boundary conditions to specify the voltage accurately here. Here I have put all the boundary conditions to be the same $f(x) = x^2$. Also I have assumed the functions to be continuous at the corners.

````python
upper_y = edge**2
lower_y, upper_x, lower_x = upper_y, upper_y, upper_y
````

````python
@jit(nopython = True,nogil = True)
def pot(p, const, iters):
    l = len(p[0])
    for _ in range(iters):
        for i in range(1,l-1):
            for j in range(1,l-1):
                if not(const[i][j]):
                    p[i][j] = 1/4 * (p[i+1][j] + p[i-1][j] + p[i][j+1] + p[i][j-1])
    return p
````

````python
%%time
iterations = 10000
p = np.zeros((300,300))
FIXED = p != 0
p[0,:]= lower_y
p[-1,:]= upper_y
p[:,0]= lower_x
p[:,-1]= upper_x

p1 = pot(p,FIXED, iterations)
````

````python
fig, axes = plt.subplots(figsize = (12,9), nrows = 1, ncols = 1)
potplot = axes.contourf(xv, yv, p1, 100)
fig.colorbar(potplot, label = "Norm Voltage")
````

````python
# @jit(nopython = True, nogil = True)
box = [[-0.25,0.25],[-0.25,0,25]]
#     return np.select([(x > box[0][0]) * (x < box[0][1]) * (y < box[1][1]) * (y > box[1][0]), (x >= box[0][1])+(x <= box[0][0])+(y >= box[1][1])+(y <= box[1][0])],(1,0))
def fixed(x, y):
    return np.select([(x > -0.25) * (x < 0.25) * (y < 0.25) * (y > -0.25), (x >= 0.25)+(x <= -0.25)+(y >= 0.25)+(y <= -0.25)],(1,0))
````

````python
block = fixed(xv, yv)
block_bool = block!=0
````

````python
p = np.zeros((300,300))
p[0,:]= lower_y
p[-1,:]= upper_y
p[:,0]= lower_x
p[:,-1]= upper_x
p[block_bool] = block[block_bool]
p2 = pot(p, block_bool, iterations)
````

````python
fig, axes = plt.subplots(figsize = (11,9), nrows = 1, ncols = 1)
potplot = axes.contourf(xv, yv, p2, 100)
fig.colorbar(potplot)
````
