```python
import numpy as np
import matplotlib.pyplot as plt
from numba import njit
import scipy.stats as stats
```

```python
@njit

def Model(theta, n):
    mu = theta[0]
    sigma = theta[1]
    x = np.zeros(n)
    t_e = 10
    t_0 = 0
    Dt = (t_e-t_0)/n
    x[0] = 1
    t = np.arange(t_0+Dt, t_e+Dt,Dt)
    for i in range(1,n):
        x[i] = x[i-1] + mu * x[i-1] * Dt + sigma * x[i-1] * np.random.normal(0,np.sqrt(Dt))
    return x,t
```
