```python
import numpy as np
import matplotlib.pyplot as plt
from numba import jit
```

```python
t = 1000000
D_r = 0.16/t
D_t = (0.22 * 10**(-6))/t
v = np.array([0,1,2,3])
v = (v*10**(-6))/t
O =np.array([0, np.pi, -np.pi])
O = (O*10**(-6))/t
```

```python
DT = 10/t
```

```python
@jit
def path():
    phi = np.zeros(t)
    x = np.zeros(t)
    y = np.zeros(t)
    for i in range(0, t-1):
        phi[i+1] = phi[i] + O[0]*DT + np.sqrt(2*D_r*DT)*np.random.normal(0,1)
        x[i+1] = x[i] + v[1]*np.cos(phi[i])*DT + np.sqrt(2*D_t*DT)*np.random.normal(0,1)
        y[i+1] = y[i] + v[1]*np.sin(phi[i])*DT + np.sqrt(2*D_t*DT)*np.random.normal(0,1)
    return x, y, phi
```

```python
x = path()[0]
y = path()[1]
plt.plot(x,y)
```
