# Stone Weierstrass Theorem

This theorem says in layman terms that any continuous function $f(x)$ over an interval $[a,b]$ can be arbitrarily approximated by a polynomial function $p(x)$

**Theorem** 

If $f(x)$ is a real function defined over $[a,b]$, there exists a sequence of real polynomials $p_n(x)$ s.t. 
$$lim_{n\rightarrow \infty}p_n(x) = f(x)$$
uniformly on $[a,b]$


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import sympy as smp
from sympy import init_printing
init_printing()
```

```python
intal = [0,1]
intv = [-1,1]
```

```python
def Q(x,n):
    return (1-x**2)**n
def f(x):
    return np.sin(np.pi *x)
def C(g,m,interval):
    h = simpsons(g,m,interval)
    return h
def simpsons(npt,n1,interval):
    h = (interval[1]-interval[0])/(npt)
    integral = 0
    for i in range(0,npt):
        t1= interval[0]+i*h
        t2 = interval[0]+(i+1)*h
        t3 = interval[0]+i*h+h/2
        m1 = Q(t1,n1)
        m2 = Q(t2,n1)
        m3 = Q(t3,n1)
        integral += h/6*(m1+m2+4*m3)
    return integral
```

```python
interval = [0,1]
npt = 1000
num = 10
x = np.linspace(interval[0],interval[1],npt)
g = np.array([Q(x,i)/C(npt,i,intv) for i in range(1,num+1)])
P_n = [sp.signal.convolve(f(x),g[i],mode = "same") for i in range(0,num)]
```

```python
fig,axes = plt.subplots(figsize = (16,16), nrows = 3,ncols = 1)
axes[0].plot(x,f(x))
axes[0].set_xlim([0,1])
axes[1].plot(x,Q(x,1))
axes[1].set_xlim([0,1])
axes[1].set_ylim([0,1])
axes[0].plot(x,P_n[8])
axes[2].set_xlim([0,1])
# axes[2].set_ylim([0,1])
```
