# The Weierstrass Approximation theorem

```python
import numpy as np
import matplotlib.pyplot as plt
from numba import jit
```

```python
interval = [0,1]
```

```python
@jit
def f(x):
    return np.e**x*np.sin(6*x)/6*x + np.cos(12*x)
def fact(k):
    fa = 1 
    for i in range(1,k+1):
        fa *= i   
    return fa
def choose(n,k):
    return fact(n)/(fact(n-k)*fact(k))
```

```python

def B(x,n):
    b = 0
    for i in range(n+1):
        b += choose(n,i)*f(i/n)*(x**i)*((1-x)**(n-i))
    return b
```

```python
npt = 1000
num = 500
x = np.linspace(interval[0],interval[1],npt)
P_n = [B(x,i) for i in range(1,num+1)]
```

```python
fig,axes = plt.subplots(figsize = (16,16),dpi =600, nrows = 1,ncols = 1)
axes.plot(x,f(x),'g')
[axes.plot(x,P_n[i],color = 'orange') for i in range(1,num)];

```
