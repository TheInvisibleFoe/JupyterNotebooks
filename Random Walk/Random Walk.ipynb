```python
import matplotlib.pyplot as plt
import numpy as np
import random as rand
import pylab
from collections import Counter
d = 1
s = (1,-1)
```

```python
steps =10000
grn = 1000
x = [[400]*steps for _ in range(grn)]
for j in range(0,grn):
    for i in range(1,steps):
        t = rand.choice(s)
        x[j][i]=(x[j][i-1] + t*d)
y = np.arange(0,steps,1)
mx = max(x[0])
mn = min(x[0])
x_1 = [[0]*(mx-mn+1) for _ in range(grn)]
y_1 = [[0]*(mx-mn+1) for _ in range(grn)]
for i in range(0,grn):
    frequency_counter = Counter(x[i])
    for key, value in frequency_counter.items():
        x_1[i].append(key)
        y_1[i].append(value)
# fig,axes = plt.subplots(figsize=(16,9),dpi=400,nrows = grn, ncols = 2,squeeze=False)
# for i in range(0,grn):
#     axes[i][0].plot(y,x[i])
#     axes[i][1].bar(x_1[i],y_1[i])
#     axes[i][0].grid()
last = np.zeros(grn)
for i in range(grn):
    last[i] = abs(x[i][-1])
plt.hist(last)
# last
```

```python
msd = [[0]*steps for _ in range(grn)]
x_sq = [[0]*steps for _ in range(grn)]
for j in range(0,grn):    
    for i in range(1,steps):
        x_sq[j] = [i**2 for i in x[j]]
        msd[j][i] = (1/(i))*(sum(x_sq[j][:i]))
        
plt.plot(y,msd[0])
```

```python
step = 1000000
s_2 = (-2,-1,1,2)
x_2 = np.zeros(step)
y_2 = np.zeros(step)
for i in range(1,step):
    t = np.random.choice(s_2)
    if abs(t) == 2:
        x_2[i]=x_2[i-1]+(t/2)*d
        y_2[i] = y_2[i-1]
    elif abs(t) == 1:
        x_2[i] = x_2[i-1]
        y_2[i]=y_2[i-1]+t*d
fig,axes = plt.subplots(figsize = (16,16))
axes.scatter(x_2[-1],y_2[-1],c='r',zorder=2)
axes.scatter(x_2[0],y_2[0],c='k',zorder=2)
axes.plot(x_2,y_2,zorder=1)
```

```python
step = 100000
s_2 = (-3,-2,-1,1,2,3)
x_2 = np.zeros(step)
y_2 = np.zeros(step)
z_2 =np.zeros(step)
for i in range(1,step):
    t = rand.choice(s_2)
    if abs(t) == 2:
        x_2[i]=x_2[i-1]+(t/2)*d
        y_2[i] = y_2[i-1]
        z_2[i] = z_2 [i-1]
    elif abs(t) == 1:
        x_2[i] = x_2[i-1]
        y_2[i]=y_2[i-1]+t*d
        z_2[i] = z_2 [i-1]
    else:
        x_2[i] = x_2[i-1]
        y_2[i] = y_2[i-1]
        z_2[i] = z_2 [i-1]+(t/3)*d
fig = plt.figure(figsize = (16,16))
axes = plt.axes(projection ='3d')
axes.annotate(x_2[-1],y_2[-1],z_2[-1],c='r',zorder=2)
axes.scatter(x_2[0],y_2[0],z_2[0],c='k',zorder=2)


axes.plot(x_2,y_2,z_2,zorder=1)

```
