# Random Walk in 2D

```python
step = 100000
s_2 = (-2,-1,1,2)
x_2 = np.zeros(step)
y_2 = np.zeros(step)
for i in range(1,step):
    t = np.random.choice(s_2)
    if abs(t) == 2:
        x_2[i]=x_2[i-1]+(t/2)*d
        y_2[i] = y_2[i-1]
    elif abs(t) == 1:
        x_2[i] = x_2[i-1]
        y_2[i]=y_2[i-1]+t*d
fig,axes = plt.subplots(figsize = (16,16))
axes.scatter(x_2[-1],y_2[-1],c='r',zorder=2)
axes.scatter(x_2[0],y_2[0],c='r',zorder=2)
# axes.axline((0,0))

axes.plot(x_2,y_2,'c',zorder=1)
```

```python
step = 1000
s_2 = (-3,-2,-1,1,2,3)
x_2 = np.zeros(step)
y_2 = np.zeros(step)
z_2 =np.zeros(step)
for i in range(1,step):
    t = rand.choice(s_2)
    if abs(t) == 2:
        x_2[i]=x_2[i-1]+(t/2)*d
        y_2[i] = y_2[i-1]
        z_2[i] = z_2 [i-1]
    elif abs(t) == 1:
        x_2[i] = x_2[i-1]
        y_2[i]=y_2[i-1]+t*d
        z_2[i] = z_2 [i-1]
    else:
        x_2[i] = x_2[i-1]
        y_2[i] = y_2[i-1]
        z_2[i] = z_2 [i-1]+(t/3)*d
fig = plt.figure(figsize = (16,16))
axes = plt.axes(projection ='3d')
axes.scatter(x_2[-1],y_2[-1],z_2[-1],c='r',zorder=2)
axes.scatter(x_2[0],y_2[0],z_2[0],c='k',zorder=2)


axes.plot(x_2,y_2,z_2,zorder=1)

```
