# Random Walk in 2D

```python
step = 10000
grn = 1
starting = [300,400]
s_2 = (-2,-1,1,2)
x_2 = np.full((grn,step),starting[0])
y_2 = np.full((grn,step),starting[1])
for i in range(0,grn):
    for j in range(1,step):
        t = np.random.choice(s_2)
        if abs(t) == 2:
            x_2[i][j]=x_2[i][j-1]+(t/2)*d
            y_2[i][j] = y_2[i][j-1]
        elif abs(t) == 1:
            x_2[i][j] = x_2[i][j-1]
            y_2[i][j]=y_2[i][j-1]+t*d
fig,axes = plt.subplots(figsize = (16,9))
# axes.axline((0,0))
axes.scatter(x_2[0][-1],y_2[0][-1],c='r',zorder=2)
axes.scatter(x_2[0][0],y_2[0][0],c='r',zorder=2)
axes.plot(x_2[0],y_2[0],'c',zorder=1)
```

```python
binwidth = 2
last_x=np.zeros(grn)
last_y=np.zeros(grn)
for i in range(grn):
    last_x[i] = abs(x_2[i][-1])
    last_y[i] = abs(y_2[i][-1])
fig2,ax = plt.subplots(figsize =(16,9),nrows = 1,ncols=2,squeeze =False)
ax[0][0].hist(last_x,bins= np.arange(min(last_x), max(last_x) + binwidth, binwidth))
ax[0][1].hist(last_y,bins= np.arange(min(last_y), max(last_y) + binwidth, binwidth))
```

# Random Walk in 3D

```python
step = 100000
s_2 = (-3,-2,-1,1,2,3)
x_2 = np.zeros(step)
y_2 = np.zeros(step)
z_2 =np.zeros(step)
for i in range(1,step):
    t = rand.choice(s_2)
    if abs(t) == 2:
        x_2[i]=x_2[i-1]+(t/2)*d
        y_2[i] = y_2[i-1]
        z_2[i] = z_2 [i-1]
    elif abs(t) == 1:
        x_2[i] = x_2[i-1]
        y_2[i]=y_2[i-1]+t*d
        z_2[i] = z_2 [i-1]
    else:
        x_2[i] = x_2[i-1]
        y_2[i] = y_2[i-1]
        z_2[i] = z_2 [i-1]+(t/3)*d
fig = plt.figure(figsize = (16,16))
axes = plt.axes(projection ='3d')
axes.scatter(x_2[-1],y_2[-1],z_2[-1],c='r',zorder=2)
axes.scatter(x_2[0],y_2[0],z_2[0],c='k',zorder=2)


axes.plot(x_2,y_2,z_2,zorder=1)

```
