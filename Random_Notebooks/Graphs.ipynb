# f(x)= 2x^2 - 8x -24

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x=np.linspace(-100,100,1000)
def f(x):
    a = 2*(x**2)-8*x-24
    return a

fig,axes = plt.subplots(figsize=(16,16),dpi=400,nrows = 2, ncols = 2)
axes[0][0].set_xlim([-10,10])
axes[0][0].set_ylim([-40,40])
axes[0][0].set_title('y = f(x)')
axes[0][0].annotate("(6,0)",(6,0))
axes[0][0].annotate("(-2,0)",(-2,0))
axes[0][0].set_xticks(np.arange(-10,10,5))
axes[0][0].set_yticks(np.arange(-40,40,10))
axes[0][0].spines['left'].set_position('center')
axes[0][0].spines['bottom'].set_position('center')
axes[0][0].plot(x,f(x),'r',label = 'y=f(x)')
axes[0][0].legend()
axes[0][0].plot(6,0,'bo')
axes[0][0].plot(-2,0,'bo')
axes[0][0].grid(True)

axes[1][0].set_xlim([-10,10])
axes[1][0].set_ylim([-40,40])
axes[1][0].set_title('y = f(|x|)')
axes[1][0].annotate("(6,0)",(6,0))
axes[1][0].annotate("(-6,0)",(-6,0))
axes[1][0].annotate("(-2,0)",(-2,0))
axes[1][0].set_xticks(np.arange(-10,10,5))
axes[1][0].set_yticks(np.arange(-40,40,10))
axes[1][0].spines['left'].set_position('center')
axes[1][0].spines['bottom'].set_position('center')
axes[1][0].plot(x,f(abs(x)),'r',lw=2,label = 'y=f(|x|)')
axes[1][0].plot(x,f(x),'g-.',lw=2,label = 'y=f(x)')
axes[1][0].plot(6,0,'bo')
axes[1][0].plot(-2,0,'bo')
axes[1][0].plot(-6,0,'bo')
axes[1][0].legend()
axes[1][0].grid(True)

axes[0][1].set_xlim([-10,10])
axes[0][1].set_ylim([-40,40])
axes[0][1].set_title('y = |f(x)|')
axes[0][1].annotate("(6,0)",(6,0))
axes[0][1].annotate("(-2,0)",(-2,0))
axes[0][1].set_xticks(np.arange(-10,10,5))
axes[0][1].set_yticks(np.arange(-40,40,10))
axes[0][1].spines['left'].set_position('center')
axes[0][1].spines['bottom'].set_position('center')
axes[0][1].plot(x,abs(f(x)),'r',lw=2,label = 'y=|f(x)|')
axes[0][1].plot(x,f(x),'g-.',lw=2,label = 'y=f(x)')
axes[0][1].plot(6,0,'bo')
axes[0][1].plot(-2,0,'bo')
axes[0][1].legend()
axes[0][1].grid(True)

axes[1][1].set_xlim([-10,10])
axes[1][1].set_title('y = |f(|x|)|')
axes[1][1].set_ylim([-40,40])
axes[1][1].annotate("(6,0)",(6,0))
axes[1][1].annotate("(-6,0)",(-6,0))
axes[1][1].annotate("(-2,0)",(-2,0))
axes[1][1].set_xticks(np.arange(-10,10,5))
axes[1][1].set_yticks(np.arange(-40,40,10))
axes[1][1].spines['left'].set_position('center')
axes[1][1].spines['bottom'].set_position('center')
axes[1][1].plot(x,abs(f(abs(x))),'r',lw=2,label = 'y=|f(|x|)|')
axes[1][1].plot(x,f(x),'g-.',lw=2,label = 'y=f(x)')
axes[1][1].plot(6,0,'bo')
axes[1][1].plot(-6,0,'bo')
axes[1][1].plot(-2,0,'bo')
axes[1][1].legend()
axes[1][1].grid(True)

    
```
