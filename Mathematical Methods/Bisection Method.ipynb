# Bisection Method

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return (x-3)*(x-1)
```

```python
def bisection(a,b,t):
    xpt = []
    ypt = []
    xpt.append(a)
    ypt.append(f(a))
    if f(a)*f(b)>0:
        print("invalid Interval")
    elif f(a)*f(b)==0:
        if f(a)==0:
            print(str(a)+" is a root of (x-3)**2 +(x-2)*(x-3) ")
            xpt.append(a)
            ypt.append(f(a))
        else:
            print(str(b)+" is a root of (x-3)**2 +(x-2)*(x-3) ")
            xpt.append(b)
            ypt.append(f(b))
    elif f(a)*f(b)<0:
        while True:
            c = (a+b)/2
            if f(c)*f(a)<0:
                xpt.append(c)
                ypt.append(f(c))
                b=c
            elif f(c)*f(b)<0:
                xpt.append(c)
                ypt.append(f(c))
                a=c
            elif f(c)==0:
                xpt.append(c)
                break
            if abs(f(c))<t:
                break
    return xpt,ypt
l,r = bisection(2,10.1,0.00000000000001)
print("This is a close to root "+str(l[-1]))
```
